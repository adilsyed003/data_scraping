```python
from bs4 import BeautifulSoup
import requests
```

```python
data = requests.get('https://en.wikipedia.org/wiki/List_of_largest_companies_in_India')
```

```python
soup = BeautifulSoup(data.text,'html')
table = soup.find_all('table',class_='wikitable sortable')[0]
print(table)
```

```python
table_head = table.find_all('th')
```

```python
new_table_head = [head.text.strip() for head in table_head]
```

```python
new_table_head
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=new_table_head)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```
